```python
class atm:

    def _init_(self):
        self.pin = None
        self.balance = 0
        self.language()
        

    def language(self):
        lang = input("Select language / ભાષા પસંદ કરો:\n1. English\n2. ગુજરાતી\n3. हिन्दी\nEnter choice: ")
        if lang == '1':
            self.start_with_english()
        elif lang == '2':
            self.start_with_gujarati()
        elif lang == '3':
            self.start_with_hindi()
        else:
            self.start_with_english()

    def start_with_english(self):
        print("Welcome to the ATM!")
        self.create_pin()
        
        while True:
            self.display_menu()
            choice = input("Please select an option (1-5): ")
            
            if choice == '1':
                if self.pin is not None:
                    self.change_pin()
                else:
                    print("Please create a PIN first")

            elif choice == '2':
                if self.pin is not None:
                    self.check_balance()
                else:
                    print("Please create a PIN first")

            elif choice == '3':
                if self.pin is not None:
                    self.deposit()
                else:
                    print("Please create a PIN first")

            elif choice == '4':
                if self.pin is not None:
                    self.withdraw()
                else:
                    print("Please create a PIN first")

            elif choice == '5':
                print("Thank you for using the ATM. Goodbye!")
                break

            else:
                print("Invalid choice. Please select a valid option.")



    def create_pin(self):
        pin = input("Generate a PIN: ")
        if len(pin) == 4:
            self.pin = pin
            print("PIN created successfully")
        else:
            print("Entered PIN is not valid, PIN must be a 4-digit number.please Enter PIN again to Generate.")
            self.create_pin()
            
    def display_menu(self):
        print("\n---- ATM Menu ----")
        print("1. Change PIN")
        print("2. Check balance")
        print("3. Deposit money")
        print("4. Withdraw money")
        print("5. Exit")
        
    def change_pin(self):
        entered_pin = input("Enter your current PIN: ")
        if entered_pin == self.pin:
            new_pin = input("Enter your new 4-digit PIN: ")
            self.pin = new_pin
            print("PIN changed successfully.")
        else:
            print("Incorrect PIN. Please try again.")

    def check_balance(self):
        entered_pin = input("Enter your PIN to check balance: ")
        if entered_pin == self.pin:
            print(f"Your current balance is: ₹{self.balance}")
        else:
            print("Incorrect PIN. Access Denied.")

    def deposit(self):
        entered_pin = input("Enter your PIN to deposit money: ")
        if entered_pin == self.pin:
            amount = float(input("Enter the amount to deposit: ₹"))
            self.balance += amount
            print(f"₹{amount} deposited successfully. New balance: ₹{self.balance}")
        else:
            print("Incorrect PIN. Access Denied.")

    def withdraw(self):
        entered_pin = input("Enter your PIN to withdraw money: ")
        if entered_pin == self.pin:
            amount = float(input("Enter the amount to withdraw: ₹"))
            if amount <= self.balance:
                self.balance -= amount
                print(f"₹{amount} withdraw successfully. New balance: ₹{self.balance}")
            else:
                print("Insufficient balance.")
        else:
            print("Incorrect PIN. Access Denied.")



    def start_with_gujarati(self):
        print("ATM માં આપનું સ્વાગત છે!")
        self.create_pin_g()
        
        while True:
            self.display_menu_g()
            choice = input("કૃપા કરીને એક વિકલ્પ પસંદ કરો (1-5): ")
            
            if choice == '1':
                if self.pin is not None:
                    self.change_pin_g()
                else:
                    print("કૃપા કરીને પહેલા એક પિન બનાવો.")

            elif choice == '2':
                if self.pin is not None:
                    self.check_balance_g()
                else:
                    print("કૃપા કરીને પહેલા એક પિન બનાવો.")

            elif choice == '3':
                if self.pin is not None:
                    self.deposit_g()
                else:
                    print("કૃપા કરીને પહેલા એક પિન બનાવો.")

            elif choice == '4':
                if self.pin is not None:
                    self.withdraw_g()
                else:
                    print("કૃપા કરીને પહેલા એક પિન બનાવો.")

            elif choice == '5':
                print("આ ATM નો ઉપયોગ કરવા બદલ આભાર. તમારો દિવસ શુભ રહે!")
                break

            else:
                print("અમાન્ય પસંદગી. કૃપા કરીને એક માન્ય વિકલ્પ પસંદ કરો.")



    def create_pin_g(self):
        print("કૃપા કરીને તમારો પિન જનરેટ કરો.")
        pin = input("તમે સેટ કરવા માંગો છો તે તમારો PIN દાખલ કરો: ")
        if len(pin) == 4:
            self.pin = pin
            print("પિન સફળતાપૂર્વક બનાવવામાં આવ્યો.")
        else:
            print("દાખલ કરેલ પિન માન્ય નથી, પિન 4-અંકનો નંબર હોવો જોઈએ. કૃપા કરીને પિન જનરેટ કરવા માટે ફરીથી પિન દાખલ કરો.")
            self.create_pin_g()
            
    def display_menu_g(self):
        print("\n---- ATM મેનુ ----")
        print("1. પિન બદલો")
        print("2. બેલેન્સ તપાસો")
        print("3. પૈસા જમા કરાવો")
        print("4. પૈસા ઉપાડો")
        print("5. બહાર નીકળો")
        
    def change_pin_g(self):
        entered_pin = input("તમારો વર્તમાન પિન દાખલ કરો: ")
        if entered_pin == self.pin:
            new_pin = input("તમારો નવો 4-અંકનો પિન દાખલ કરો: ")
            self.pin = new_pin
            print("પિન સફળતાપૂર્વક બદલાયો!")
        else:
            print("ખોટો પિન. પિન બદલી શકાતો નથી.")

    def check_balance_g(self):
        entered_pin = input("બેલેન્સ ચેક કરવા માટે તમારો પિન દાખલ કરો: ")
        if entered_pin == self.pin:
            print(f"તમારું વર્તમાન બેલેન્સ છે: ₹{self.balance}")
        else:
            print("ખોટો પિન. ઍક્સેસ નકારવામાં આવી.")

    def deposit_g(self):
        entered_pin = input("પૈસા જમા કરવા માટે તમારો પિન દાખલ કરો: ")
        if entered_pin == self.pin:
            amount = float(input("જમા કરાવવાની રકમ દાખલ કરો: ₹"))
            self.balance += amount
            print(f"₹{amount} સફળતાપૂર્વક જમા કરાવ્યું. નવું બેલેન્સ: ₹{self.balance}")
        else:
            print("ખોટો પિન. ઍક્સેસ નકારવામાં આવી.")

    def withdraw_g(self):
        entered_pin = input("પૈસા ઉપાડવા માટે તમારો પિન દાખલ કરો: ")
        if entered_pin == self.pin:
            amount = float(input("ઉપાડવાની રકમ દાખલ કરો: ₹"))
            if amount <= self.balance:
                self.balance -= amount
                print(f"₹{amount} સફળતાપૂર્વક ઉપાડ. નવું બેલેન્સ: ₹{self.balance}")
            else:
                print("અપૂરતું બેલેન્સ.")
        else:
            print("ખોટો પિન. ઍક્સેસ નકારવામાં આવી.")



    def start_with_hindi(self):
        print("एटीएम में आपका स्वागत है!")
        self.create_pin_h()
        
        while True:
            self.display_menu_h()
            choice = input("कृपया एक विकल्प चुनें (1-5): ")
            
            if choice == '1':
                if self.pin is not None:
                    self.change_pin_h()
                else:
                    print("कृपया पहले एक पिन बनाएं।")

            elif choice == '2':
                if self.pin is not None:
                    self.check_balance_h()
                else:
                    print("कृपया पहले एक पिन बनाएं।")

            elif choice == '3':
                if self.pin is not None:
                    self.deposit_h()
                else:
                    print("कृपया पहले एक पिन बनाएं।")

            elif choice == '4':
                if self.pin is not None:
                    self.withdraw_h()
                else:
                    print("कृपया पहले एक पिन बनाएं।")

            elif choice == '5':
                print("एटीएम का उपयोग करने के लिए धन्यवाद। आपका दिन मंगलमय हो!")
                break

            else:
                print("अमान्य विकल्प. कृपया मान्य विकल्प चुनें।")



    def create_pin_h(self):
        pin = input("पिन जनरेट करें: ")
        if len(pin) == 4:
            self.pin = pin
            print("पिन सफलतापूर्वक बनाया गया।")
        else:
            print("दर्ज किया गया पिन मान्य नहीं है, पिन 4 अंकों का होना चाहिए। कृपया जनरेट करने के लिए पिन पुनः दर्ज करें।")
            self.create_pin_h()
            
    def display_menu_h(self):
        print("\n---- एटीएम मेनू ----")
        print("1. पिन बदलें")
        print("2. शेष राशि की जाँच करें")
        print("3. पैसे जमा करो")
        print("4. पैसे निकाले")
        print("5. बाहर निकलना")
        
    def change_pin_h(self):
        entered_pin = input("अपना वर्तमान पिन दर्ज करें: ")
        if entered_pin == self.pin:
            new_pin = input("अपना नया 4-अंकीय पिन दर्ज करें: ")
            self.pin = new_pin
            print("पिन सफलतापूर्वक बदला गया।")
        else:
            print("ग़लत पिन। कृपया पुनः प्रयास करें।")

    def check_balance_h(self):
        entered_pin = input("शेष राशि जांचने के लिए अपना पिन दर्ज करें: ")
        if entered_pin == self.pin:
            print(f"आपकी वर्तमान शेष राशि है: ₹{self.balance}")
        else:
            print("ग़लत पिन। प्रवेश निषेध।")

    def deposit_h(self):
        entered_pin = input("पैसा जमा करने के लिए अपना पिन दर्ज करें: ")
        if entered_pin == self.pin:
            amount = float(input("जमा करने हेतु राशि दर्ज करें: ₹"))
            self.balance += amount
            print(f"₹{amount} सफलतापूर्वक जमा किया गया। नया शेष: ₹{self.balance}")
        else:
            print("ग़लत पिन। प्रवेश निषेध।")

    def withdraw_h(self):
        entered_pin = input("पैसे निकालने के लिए अपना पिन दर्ज करें: ")
        if entered_pin == self.pin:
            amount = float(input("निकासी हेतु राशि दर्ज करें: ₹"))
            if amount <= self.balance:
                self.balance -= amount
                print(f"₹{amount} सफलतापूर्वक निकासी करें। नया बैलेंस: ₹{self.balance}")
            else:
                print("अपर्याप्त शेषराशि।")
        else:
            print("ग़लत पिन। प्रवेश निषेध।")
```
